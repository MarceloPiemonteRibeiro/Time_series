```python
# average temperature per day of the week
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
temperatura_por_dia_da_semana
```

```python
# order
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana
```

```python
fig=plt.figure(figsize=(5,4))
eixo=fig.add_axes([0,0,1,1])
# axis X:
indice=range(len(temperatura_por_dia_da_semana))
eixo.bar(indice,temperatura_por_dia_da_semana)
eixo.set_title('Temperature per day of the week', fontsize=15, pad=10)
eixo.set_xlabel('Week day', fontsize=15)
eixo.set_ylabel('Average temp', fontsize=15)
```

```python
# change colors
fig=plt.figure(figsize=(5,4))
eixo=fig.add_axes([0,0,1,1])
# axis X:
indice=range(len(temperatura_por_dia_da_semana))
cores = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']

eixo.bar(indice,temperatura_por_dia_da_semana, color=cores)
eixo.set_title('Temperature per week day', fontsize=15, pad=10)
eixo.set_xlabel('Week day', fontsize=15)
eixo.set_ylabel('Average temperature', fontsize=15)
# change axis legends
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
# add border
fig=plt.figure(figsize=(5,4))
eixo=fig.add_axes([0,0,1,1])
# axis X:
indice=range(len(temperatura_por_dia_da_semana))
cores = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']

eixo.bar(indice,temperatura_por_dia_da_semana, color=cores, edgecolor='black')
eixo.set_title('Temperatura per week day', fontsize=15, pad=10)
eixo.set_xlabel('Week day', fontsize=15)
eixo.set_ylabel('Average tempearature', fontsize=15)

eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

## Scatter plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
df.info()
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Scatter plot', fontsize=25, pad=15)
eixo.set_xlabel('lenght', fontsize=15)
eixo.set_ylabel('width', fontsize=15)
```

```python
# make the axis legends bigger
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Scatter plort', fontsize=25, pad=15)
eixo.set_xlabel('Lenght', fontsize=15)
eixo.set_ylabel('Width', fontsize=15)
eixo.tick_params(labelsize=15)
```

```python
# color each category
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

# create categories
cores={'Iris-setosa':'r','Iris-versicolor':'b', 'Iris-virginica':'g'}

# create scatter for each category
for especie in df['espécie'].unique():
    tmp=df[df['espécie']==especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                color=cores[especie])

eixo.set_title('Scatter plor', fontsize=25, pad=15)
eixo.set_xlabel('Lenght', fontsize=15)
eixo.set_ylabel('Width', fontsize=15)
eixo.tick_params(labelsize=15)

# Colors' legend
eixo.legend(cores, fontsize=20)
```

```python
# add markers
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores={'Iris-setosa':'r','Iris-versicolor':'b', 'Iris-virginica':'g'}
# add markers for each category
marcadores = {'Iris-setosa': 'x', 'Iris-versicolor': 'o', 'Iris-virginica': 'v'}

# create scatter for each category
for especie in df['espécie'].unique():
    tmp=df[df['espécie']==especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                color=cores[especie], marker=marcadores[especie], s=100)
# marker and its size, s

eixo.set_title('Scatter plot', fontsize=25, pad=15)
eixo.set_xlabel('Length', fontsize=15)
eixo.set_ylabel('Width', fontsize=15)
eixo.tick_params(labelsize=15)

# Colors' legend
eixo.legend(cores, fontsize=20)
```

```python
# add lines indicating median and average

fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Histogram', fontsize=15, pad=10)
eixo.set_xlabel('Length', fontsize=15)
eixo.grid(True)

# adding \ median and average display its correpondents grec symbols
eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
               xy=(4.5, 20), fontsize=20)

eixo.axvline(mu, color = 'k', linestyle='--')
# calculate average
eixo.annotate('average', xy=(mu-1.3, 28), fontsize=20)
# eixo annotate add the line, xy displays where the latter is placed

eixo.axvline(df['comprimento_pétala'].median(), color='g', linestyle='--')
# calculate mediand
eixo.annotate('median', xy=(df['comprimento_pétala'].median(), 31), fontsize=20)
# place median
```
